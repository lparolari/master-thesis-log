```python
import pandas as pd
```

```python
filename = 'referit_train_word_freq_only.csv'

df_word_freq = pd.read_csv(filename)
```

```python
df_word_freq.head(n=2)
```

```python
n_query = df_word_freq['freq'].sum()

print('number of queries =', n_query)
```

## Top-20 words sorted by frequency DESC

```python
k = 20

df = df_word_freq.copy()
df = df.sort_values(by=['freq'], ascending=False)

print(df.head(n=k)['freq'].sum())
df.head(n=k)
```

## Importance of 'a' and 'the' over all words

```python
df = df_word_freq.copy()

df = df.sort_values(by=['freq'], ascending=False)
df_a_the = df.loc[(df['word'] == 'a') | (df['word'] == 'the')]

print('percentage of \'a\' and \'the\' among all words =', df_a_the['freq'].sum() / df['freq'].sum() * 100)
# print(df.loc[(df['class' == 'a']) | (df['class'] == 'the')] #['freq'])#.sum() / df['freq'].sum() * 100)
```

## Top-20 words sorted by frequency ASC

```python
df = df_word_freq.copy()
df = df.sort_values(by=['freq'], ascending=True)

df.head(n=20)
```

## Number of classes by frequency

```python
k = 10

df = df_word_freq.copy()
df = df.groupby(['freq']).count()
df = df.sort_values(by=['word'], ascending=False)

print(f"number of unique words = {len(df_word_freq.index)}")
print(f"number of unique words top-{k} = {df['word'][:k].sum()}")
```

```python
df.head(n=k)
```

but what is the importante of this low-frequency classes?

```python
df = df_word_freq.copy()
df = df.loc[(df['freq'] <= 9) | (df['freq'] == 11)]

freq_top_10_by_number_of_classes_desc = df["freq"].sum()

freq_top_10_by_number_of_classes_desc_over_queries_percent = (freq_top_10_by_number_of_classes_desc / n_query) * 100

print('sum of the frequency of top-10 class with low frequency sorted by population =', freq_top_10_by_number_of_classes_desc)
print('percent among total number of query of the frequency of top-10 class with low frequency sorted by population =', freq_top_10_by_number_of_classes_desc_over_queries_percent)
```

# Class frequency

```python
filename_class = 'flickr_train_class_freq.csv'

df_class_freq = pd.read_csv(filename_class)
df_class_freq.head()
```

```python
df = df_class_freq.copy()

df = df.loc[df['class'].isin(['a'])]#, 'Aladdin', 'Workers', 'Bystanders', 'Sightseers'])]
df = df.loc[:, (df != 0).any(axis=0)]
df = df.transpose()
df = df.iloc[1:]
df = df.set_axis(['freq'], axis=1, inplace=False)
df = df.sort_values(by='freq', ascending=False)

df.head(n=20)
```
